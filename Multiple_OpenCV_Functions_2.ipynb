### Importing necessary libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Performing Morphological operations

```python
img = cv2.imread("images/edge.png")

kernel1 = np.ones((10,10),np.uint8)
kernel2 = np.ones((10,10),np.uint8)

img1 = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel1)
img2 = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel2)
img3 = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel2)
img4 = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel2)
```

```python
img = cv2.imread("images/edge.png")
plt.imshow(img)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img3)
```

```python
plt.imshow(img4)
```

### Histogram Equilization

```python
img = cv2.imread("images/hist.jpg",0)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
equ = cv2.equalizeHist(img)
res = np.hstack((img,equ)) #stacking images side-by-side
plt.imshow(res)
```

### Writing Text over Image

```python
img = np.zeros((500,500,3),np.uint8)
plt.imshow(img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img , "OpenCV" , (100,250) , font , 3 , (255,255,255) , 2 , cv2.LINE_AA )
plt.imshow(img)
```

### Drawing circle and rectangle

```python
img = np.zeros((500,500,3),np.uint8)
cv2.circle(img,(250,250),100,(0,0,255),-1)
plt.imshow(img)
```

```python
img = np.zeros((500,500,3),np.uint8)
cv2.rectangle(img,(100,100),(400,400),(0,255,255),10)
plt.imshow(img)
```

### Bitwise operations

```python
img1 = cv2.imread("images/edge.png")
img2 = cv2.imread("images/edge2.png")
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
bit_and = cv2.bitwise_and(img2, img1)
bit_or = cv2.bitwise_or(img2, img1)
bit_xor = cv2.bitwise_xor(img1, img2)
bit_not = cv2.bitwise_not(img1)
bit_not2 = cv2.bitwise_not(img2)
```

```python
plt.imshow(bit_and)
```

```python
plt.imshow(bit_or)
```

```python
plt.imshow(bit_xor)
```

```python
plt.imshow(bit_not)
```

```python
plt.imshow(bit_not2)
```

### Blending images

```python
img1 = cv2.imread("images/edge.png")
img2 = cv2.imread("images/edge2.png")
dst = cv2.addWeighted(img1,0.8,img2,0.3,0)
plt.imshow(dst)
```
